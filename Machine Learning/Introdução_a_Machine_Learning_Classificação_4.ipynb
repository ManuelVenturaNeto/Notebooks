```python
import pandas as pd
import numpy as np
```

```python
url = 'https://gist.githubusercontent.com/guilhermesilveira/4d1d4a16ccbf6ea4e0a64a38a24ec884/raw/afd05cb0c796d18f3f5a6537053ded308ba94bf7/car-prices.csv'
```

```python
df = pd.read_csv(url)
df.head()
```

```python
df = df[ ['mileage_per_year', 'model_year', 'price', 'sold'] ]
df
```

```python
renomear = {
    'mileage_per_year' : 'milhas_ano',
    'model_year' : 'ano',
    'price' : 'preco',
    'sold' : 'vendido'
}

df = df.rename(columns = renomear)
df
```

```python
df['vendido']
```

```python
trocar = {
    'yes' : 1,
    'no' : 0
}

df['vendido'] = df.vendido.map(trocar)
df.head()
```

```python
df.info()
```

```python
from datetime import datetime


#ano_atual = 2018
ano_atual = datetime.today().year
df['idade_modelo'] = ano_atual - df.ano
df.head()
```

```python
df['km_ano'] = df.milhas_ano * 1.60934
df.head()
```

```python
df = df.drop(columns = ['milhas_ano', 'ano'], axis=1) #axis 1 refere a colula e axis 0 refere a linha
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC, SVC
from sklearn.metrics import accuracy_score

x = df[['preco', 'idade_modelo', 'km_ano']]
y = df[['vendido']]

SEED = 5
np.random.seed(SEED)
[x_train, x_test, y_train, y_test] = train_test_split (x, y, test_size=0.25, stratify=y)

print("Treinaremos com %d elementos e testaremos com %d elementos" % (len(x_train), len(x_test)))

modelo = LinearSVC()
modelo.fit(x_train, y_train.values.ravel())
previsao = modelo.predict(x_test)

acuracia = accuracy_score(y_test, previsao) * 100
print("A acurácia foi %.2f%%" % acuracia)
```
